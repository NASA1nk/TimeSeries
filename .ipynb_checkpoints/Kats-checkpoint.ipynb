```python
from datetime import datetime, timedelta
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

from kats.consts import TimeSeriesData
from kats.models.prophet import ProphetModel, ProphetParams

# 单变量数据，一定要包含时间列time
air_passengers_df = pd.read_csv('katsdata/air_passengers.csv',header=0,names=["time", "value"])
air_passengers_df.head()
# print(air_passengers_df.describe())
```

```python
air_passengers_df.plot()
```

```python
# convert to TimeSeriesData object
air_passengers_ts = TimeSeriesData(air_passengers_df)
# create a model param instance
params = ProphetParams(seasonality_mode='multiplicative') # additive mode gives worse results
# create a prophet model instance
m = ProphetModel(air_passengers_ts, params)
# fit model simply by calling m.fit()
m.fit()
# make prediction for next 30 month
fcst = m.predict(steps=30, freq="MS")
fcst.tail()
```

```python
# 画图展示
m.plot()
```

```python
from kats.consts import TimeSeriesIterator
from kats.detectors.cusum_detection import CUSUMDetector

df = pd.read_csv('./detectiondata.csv',header=0,names=['time','value'])
df.head()
```

```python
train = df[:500]
print(train)
dts = TimeSeriesData(train)
dts.plot(cols=['value'])
```

```python
from kats.detectors.outlier import OutlierDetector
ret = OutlierDetector(data=dts, decomp='additive')
ret.detector()

print(ret.outliers[0])
```

```python
# LSTM 模型
from kats.models.lstm import LSTMModel, LSTMParams

# 创建参数实例
params = LSTMParams(hidden_size=16, time_window=24, num_epochs=100)

# 创建模型实例
lstm = LSTMModel(data=dts, params=params)

# 模型训练
lstm.fit()

# 模型预测
fcst = lstm.predict(steps=30)

# 预测结果可视化展示
lstm.plot()

# 展示模型预测结果
#预测结果是一个pd.DataFrame
#fcst:预测值，fcst_lower预测下界，fcst_upper预测上界
fcst.head()
```
