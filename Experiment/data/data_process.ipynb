```python

# 查看KpiAnomaly目录下的数据
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import glob

files = glob.glob("./2018AIOpsData/*.csv")
cnt = len(files)
print(cnt)

for i, file in enumerate(files):
    df = pd.read_csv(file)
    name = file.split('/')[-1][:-4]
    marker = []
    for index, row in enumerate(df.loc[:,'label']):
        if row == 1:
            marker.append(index)
    fig, ax = plt.subplots(1, 1, figsize=(20, 10))
    fig.patch.set_facecolor('white')
    x = df["timestamp"]
    y = df["value"]
    ax.plot(x, y, marker='o', markerfacecolor = 'r', markevery=marker)
    ax.set_xlabel("timestamp")
    ax.set_ylabel("value")
    ax.set_title(f"{name}")
    plt.savefig(f'./2018AIOpsData/img/{name}.png')
```
