```python
# 标记有标签的验证集
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('./data/metric_detection/evaluation_data_with_label.csv')

# markevery标记点需要传x轴的索引
marker = []
for index, row in enumerate(df.loc[:,'label']):
    if row == 1:
        marker.append(index)
# print(marker)
fig, ax = plt.subplots(1, 1, figsize=(20, 10))
fig.patch.set_facecolor('white')
x = df["timestamp"]
y = df["value"]
ax.plot(x, y, marker='o', markerfacecolor = 'r', markevery=marker)
ax.set_xlabel("timestamp")
ax.set_ylabel("value")
ax.set_title("evaluation_data_with_label")
# plt.savefig('./data/evaluation_data_with_label.pdf')
plt.show()

```

```python
# 查看metric_detection目录下所有的数据
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import glob

# 使用通配符遍历目录下所有csv文件
files = glob.glob("../data/metric_forecast/concept_drift_data/*.csv")
# print(files)
cnt = len(files)
print(cnt)

f = plt.figure(figsize=(30, 20), dpi=800)
f.patch.set_facecolor('white')

for i, file in enumerate(files):
    # 添加子图(m,n,index)根据cnt划分
    ax = f.add_subplot(3, 1 ,i+1)
    df = pd.read_csv(file)
    ax.plot(df["timestamp"], df["value"])
    ax.set_xlabel("timestamp")
    ax.set_ylabel("value")
    ax.set_title(f'({i+1})')
plt.savefig('./data/metric_forecast/concept_drift_data.pdf')
# plt.show()
```
