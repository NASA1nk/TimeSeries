```python
# 
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('./data/evaluation_data_with_label.csv')

# markeveryx
marker = []
for index, row in enumerate(df.loc[:,'label']):
    if row == 1:
        marker.append(index)
# print(marker)
fig, ax = plt.subplots(1, 1, figsize=(20, 10))
fig.patch.set_facecolor('white')
x = df["timestamp"]
y = df["value"]
ax.plot(x, y, marker='o', markerfacecolor = 'r', markevery=marker)
ax.set_xlabel("timestamp")
ax.set_ylabel("value")
ax.set_title("evaluation_data_with_label")
plt.savefig('./data/evaluation_data_with_label.pdf')


```
