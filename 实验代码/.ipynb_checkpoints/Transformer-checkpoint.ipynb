```python
import torch
import torch.nn as nn
import numpy as np
import time
import math
from matplotlib import pyplot

# 生成相同随机数
SEED=42
np.random.seed(SEED)
torch.manual_seed(SEED)

# 测试cuda是否可用
# print(torch.cuda.is_available())

# 返回gpu数量
print(torch.cuda.device_count())
# 返回gpu名字，设备索引默认从0开始
print(torch.cuda.get_device_name(0))

# 将数据转移到GPU
# device = torch.device('cuda')
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# 位置编码,得到token的绝对位置信息和相对位置信息
# 构造一个跟输入embedding维度一样的矩阵,然后跟输入embedding相加得到multi-head attention的输入

# 无学习参数的位置编码
class PositionalEncoding(nn.Module):
    
#     d_model表示向量维度,max_len表示最大长度为5000,一般是200
    def __init__(self, d_model, max_len=5000):
        super(PositionalEncoding, self).__init__()  
#         二维矩阵pe,行表示个数len,列表示维度d_model
        pe = torch.zeros(max_len, d_model)
    
#         unsqueeze()用于增加维度,返回的张量与输入张量共享内存,改变其中一个的内容也会改变另一个
#         1表示在第二个维度(从0开始),即列方向上增加一个维度,将5000变成(5000*1)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
#         2是步长,计算公式 e^(2i*-log10000/d_model)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
#         填充pe矩阵
#         偶数列，正弦编码
        pe[:, 0::2] = torch.sin(position * div_term)
#         奇数列，余弦编码
        pe[:, 1::2] = torch.cos(position * div_term)
    
#         0表示在第一个维度前增加一个维度,将(5000*1)变成(1*5000*1)
#         第一个维度用于接受batch_size参数
#         transpose()用于对矩阵进行转置,(0,1)表示转置0维和1维,将(1*5000*1)变成(5000*1*1)
        pe = pe.unsqueeze(0).transpose(0, 1)
    
#         pe.requires_grad = False
#         存到内存中的一个常量，模型保存和加载的时候可以写入和读出,可以在forward函数中使用
#         pytorch一般情况下将网络中的参数保存成OrderedDict形式
#         网络参数包括2种,一种是模型中各种module含的参数,即nn.Parameter,也可以在网络中定义其他的nn.Parameter参数,另外一种是buffer
#         nn.Parameter在每次optim.step会得到更新，buffer不会被更新,buffer的更新在forward中
        self.register_buffer('pe', pe)
    
#     前向传播,将embedding加上position encoding
    def forward(self, x):
#         x的形状是(batch_len,batch_size,d_model)
        return x + self.pe[:x.size(0), :]
    
    
    
# 有学习参数的位置编码
class LearnedPositionEncoding(nn.Embedding):
    def __init__(self,d_model, dropout = 0.1,max_len = 5000):
        super().__init__(max_len, d_model)
        self.dropout = nn.Dropout(p = dropout)
 
    def forward(self, x):
        weight = self.weight.data.unsqueeze(1)
        x = x + weight[:x.size(0),:]
        return self.dropout(x)
```

```python
# number of input steps
input_window = 100 
# number of prediction steps, in this model its fixed to one
output_window = 1 
batch_size = 10 

# if window is 100 and prediction step is 1
# 目的是输入in = seq->[0..99],输出label = target->[1..100]

# 调用:train_sequence = create_inout_sequences(train_data,input_window)
def create_inout_sequences(input_data, tw):
    inout_seq = []
#     input_data：原始整个训练集，tw：input_window = 100
    L = len(input_data)
#     2000-100
    for i in range(L-tw):
#         训练集的(i,i+100)部分
        train_seq = input_data[i:i+tw]
#         训练集的(i+1,i+100+1)部分           
        train_label = input_data[i+output_window:i+tw+output_window]
#         处理训练集：[([0,100],[1,101]),([1,101],[2,102])...]
        inout_seq.append((train_seq ,train_label))
#     转换成tensor
    return torch.FloatTensor(inout_seq)


# 读取数据
def get_data():
    # construct a littel toy dataset
    # time = np.arange(0, 400, 0.1)    
    # amplitude = np.sin(time) + np.sin(time*0.05) +np.sin(time*0.12) *np.random.normal(-0.2, 0.2, len(time))

    from sklearn.preprocessing import MinMaxScaler

#     导入数据
    import pandas as pd
    from pandas import read_csv
#     header=0,使用数据文件的第一行作为列名称,将第一列作为索引
    series = read_csv('/data/minutedata.csv', header=0, index_col=0, parse_dates=True, squeeze=True)
#     对数据做归一化处理,提升模型的收敛速度,提升模型的精度
#     初始化MinMaxScaler在(-1,1)之间
    scaler = MinMaxScaler(feature_range=(-1, 1)) 
#     amplitude指序列振幅
#     使用scaler归一化数据

#     reshape用于更改数据的行列数,先(-1, 1)变为一列,归一行后再(-1)变为一行
    amplitude = scaler.fit_transform(series.to_numpy().reshape(-1, 1)).reshape(-1)
#     反归一化:reamplitude = scaler.inverse_transform(amplitude.reshape(-1, 1)).reshape(-1)

#     将样本划分为训练集和测试集
    sampels = 2000
    train_data = amplitude[:sampels]
    test_data = amplitude[sampels:]

#     convert our train data into a pytorch train tensor
#     view(-1)变成一行
#     train_tensor = torch.FloatTensor(train_data).view(-1)
    
    train_sequence = create_inout_sequences(train_data,input_window)
    train_sequence = train_sequence[:-output_window]  

#     test_data = torch.FloatTensor(test_data).view(-1) 
    test_data = create_inout_sequences(test_data,input_window)
    test_data = test_data[:-output_window]
    
#     把tensor数据放到GPU上,在GPU上运行
    return train_sequence.to(device),test_data.to(device)
```

```python
# torch.nn.Module是所有NN的基类

class TransAm(nn.Module):
    def __init__(self,feature_size=250,num_layers=1,dropout=0.1):
        super(TransAm, self).__init__()
        self.model_type = 'Transformer'
        self.src_mask = None
        
#         编码器
#         维度d_model = 250
        self.pos_encoder = PositionalEncoding(feature_size)
#         encoder_layer:10个头(默认8个),dropout=0.1(默认),FNN默认维度2048,激活函数默认是ReLU
#         CLASS torch.nn.TransformerEncoderLayer(d_model, nhead, dim_feedforward=2048, dropout=0.1, activation='relu')
        self.encoder_layer = nn.TransformerEncoderLayer(d_model=feature_size, nhead=10, dropout=dropout)
#         Encoder:只有一层encoder_layer
        self.transformer_encoder = nn.TransformerEncoder(self.encoder_layer, num_layers=num_layers) 
    
#         解码器
#         使用全连接层代替了decoder， 也可以加一下Transformer的decoder试试效果
        self.decoder = nn.Linear(feature_size,1)
        self.init_weights()

    def init_weights(self):
        initrange = 0.1    
        self.decoder.bias.data.zero_()
        self.decoder.weight.data.uniform_(-initrange, initrange)

    def forward(self,src):
        if self.src_mask is None or self.src_mask.size(0) != len(src):
            device = src.device
            mask = self._generate_square_subsequent_mask(len(src)).to(device)
            self.src_mask = mask

        src = self.pos_encoder(src)
        output = self.transformer_encoder(src,self.src_mask)
        output = self.decoder(output)
        return output


#     为输入序列生成一个square mask(方阵),屏蔽的位置填充float('-inf'),未屏蔽的位置填充float(0.0)
    def _generate_square_subsequent_mask(self, sz):

#         torch.ones(n, m)返回一个n*m的tensor
#         torch.triu(input, diagonal=0, out=None) → Tensor,返回矩阵上三角部分,其余部分定义为0
#         其中input是一个tensor,diagonal为空则保留输入矩阵主对角线与主对角线以上的元素
#         和1判断将数字转换为True和False
        mask = (torch.triu(torch.ones(sz, sz)) == 1).transpose(0, 1)

#         masked_fill(mask, value) → tensor
#         在mask值为1的位置处用value填充,mask的元素个数需要和tensor相同,但尺寸可以不同
        mask = mask.float().masked_fill(mask == 0, float('-inf')).masked_fill(mask == 1, float(0.0))
        return mask
```

```python
batch_size = 10

# 把源数据细分为长度为batch_size的块,生成模型的输入序列和目标序列
# 调用:data, targets = get_batch(train_data, i,batch_size)
def get_batch(source, i, batch_size):
#     数据从i开始
    seq_len = min(batch_size, len(source) - 1 - i)
    data = source[i:i+seq_len]    
#     对tensors沿指定维度拼接，返回一维的tensor
#     source = [([0,100],[1,101]),([1,101],[2,102])...],取出item[0],即train_seq = [[0,100],[1,101]...]
#     chunk(n,dim)可以对tensor在指定维度上(dim = 0行,dim = 1列)分为n块,返回一个tensor list
    input = torch.stack(torch.stack([item[0] for item in data]).chunk(input_window,1)) 
#     取出item[1],即train_label = [[1,101],[2,102]...]
    target = torch.stack(torch.stack([item[1] for item in data]).chunk(input_window,1))
    return input, target
```
